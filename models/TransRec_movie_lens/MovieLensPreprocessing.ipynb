```python
import glob
import pandas as pd
from collections import defaultdict
from datetime import datetime
import pandas as pd
```

```python
# config
dataset_name = 'MovieLens100k'
filename     = '../../data/' + dataset_name + '/ratings.csv'
df = pd.read_csv(filename, sep=',', header=None,
                names=['user_id', 'item_id', 'rating', 'time'], index_col=False)
```

```python
usermap = dict()
usernum = 0
itemmap = dict()
itemnum = 0
User = dict()
len(df)
```

```python
for i in range(len(df)):
    if df['user_id'][i] in usermap: userid = usermap[df['user_id'][i]]
    else:
        userid = usernum
        usernum += 1
        usermap[df['user_id'][i]] = userid
        User[userid] = []
    if df['item_id'][i] in itemmap: itemid = itemmap[df['item_id'][i]]
    else:
        itemid = itemnum
        itemnum += 1
        itemmap[df['item_id'][i]] = itemid    
    User[userid].append([df['time'][i],itemid])
```

```python
for userid in User.keys():
#     #User[userid].sort()
    User[userid].sort(key=lambda x: x[0])


# In[3]:

itemmap
```

```python
import numpy as np
dataset=[User,IIG,usermap,itemmap,usernum,itemnum]
np.save('meta_'+dataset_name+'.npy',dataset)
```
